# Creacion de diccionario

```python
import pandas as pd
import matplotlib.pyplot as plt

path="mcc_codes.csv"
data1 = pd.read_csv(path, sep=",", header=0, dtype={'mcc':str})

path="Resumen_mccgs.csv"
data2 = pd.read_csv(path, sep=",", header=0, dtype={'mcc':str,'mccg':str})
data2 = data2.fillna("Non_name")
```

```python
data = data1.join(data2.set_index('mcc'), on='mcc')
data.head(2)
```

# Ejemplo usando data de SLG

```python
import psycopg2
import sys
import pprint
from datetime import date, timedelta
import pandas as pd
```

```python
 #Connect to RedShift
conn_string = "dbname='' port='' user='' password='' host=''";
print ("Connecting to database ->{}".format(conn_string))
conn = psycopg2.connect(conn_string);
```

```python
cursor = conn.cursor();
#Captures Column Names 
column_names = [];
query = """
SELECT *
FROM transaction
WHERE merchant_district = 'La Victoria'
"""
muestra = pd.read_sql(query, conn)
muestra['mcc'] = muestra['mcc'].astype(str) 
```

```python
muestra.head(3)
```

```python
mm = muestra.join(data.set_index('mcc'), on='mcc', how='left', rsuffix='_right')
mm.head()
```

## Grafica

```python
y = list(mm.groupby(['Categoria_M']).count()['client_id'])
x = list(mm.groupby(['Categoria_M']).count()['amount_sol'] .index)

fig1, ax1 = plt.subplots()
ax1.pie(y, labels=x, autopct='%1.1f%%', shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
